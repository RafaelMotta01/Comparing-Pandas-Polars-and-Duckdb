```python
import pandas as pd
import os
import time
```

```python
# record start time
start = time.time()

# Read csv
df_app = pd.read_csv('Google-Playstore.csv')

# record end time
end = time.time()
```

```python
print("The time of execution of pandas read_csv is:",(end-start) * 10**3, "ms")
```

```python
# Selecting the desire variables
df2 = df_app[['App Name', 'Category', 'Rating', 'Installs', 'Free', 'Developer Id', 'Released', 'Last Updated']]
df2['Installs'] = df2['Installs'].astype('str')
```

```python
#Config max_rows displayed
pd.options.display.max_rows = 100
```

```python
# Maximum number of installs
df2['Installs'].unique()
```

```python
# Since the variable "Installs" dont show especific numbers, only agreggations, 
# we wont be able to find an exactly most downloaded app
df2[(df2.Installs == '10,000,000,000+') & (df2.Free == True)]
```

```python
# Since the only app with 10,000,000,000+ its the google play itself we can disconsider it 
# and use the second largest amount of downloads
df2[(df2.Installs == '5,000,000,000+') & (df2.Free == True)]
```

```python
#New pandas dataframe already filtred
df3 = df2[(df2.Installs == '5,000,000,000+') & (df2.Free == True)]
```

```python
#Group by "Category"
df3[["Category", "Installs"]].groupby("Category").count().sort_values(by="Installs", ascending=False)
```
